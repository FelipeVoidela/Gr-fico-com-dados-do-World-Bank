```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Metadata_Country_API_SH.DYN.MORT_DS2_en_csv_v2_526059.csv")
```

### Gráfico de Linha 


Vamos criar um gráfico de linha que mostre a contagem de países por grupo de renda ao longo das regiões.

```python
income_counts = df['IncomeGroup'].value_counts()

plt.figure(figsize=(10,6))
income_counts.plot(kind='line')
plt.title('Contagem de Países por Grupo de Renda')
plt.xlabel('Grupos de Renda')
plt.ylabel('Número de Países')
plt.grid(True)
plt.show()
```

## Gráfico de Barras

Vamos criar um gráfico de barras que mostre a distribuição de países por região.

```python
region_counts = df['Region'].value_counts()

plt.figure(figsize=(12,8))
region_counts.plot(kind='bar')
plt.title('Distribuição de Países por Região')
plt.xlabel('Região')
plt.ylabel('Número de Países')
plt.show()
```

## Histograma

Vamos criar um histograma para visualizar a distribuição dos países pelos grupos de renda.

```python
plt.figure(figsize=(10,6))
plt.hist(df['IncomeGroup'].dropna(), bins=10, edgecolor='k')
plt.title('Distribuição de Países pelos Grupos de Renda')
plt.xlabel('Grupo de Renda')
plt.ylabel('Frequência')
plt.show()
```

## Gráfico de Pizza 

Vamos criar um gráfico de pizza para mostrar a proporção de países em cada grupo de renda.

```python
plt.figure(figsize=(10,6))
income_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Proporção de Países por Grupo de Renda')
plt.ylabel('')
plt.show()
```

## Gráfico de Dispersão (Scatter)

Vamos criar um gráfico de dispersão para visualizar a relação entre regiões e grupos de renda.

```python
df = pd.read_csv("API_SH.DYN.MORT_DS2_en_csv_v2_526059.csv", skiprows=4, skipfooter=2, engine='python')


columns_to_remove = ['Country Code', 'Indicator Name', 'Indicator Code', 'Last Updated Date', 'Unnamed: 68']
columns_to_remove = [col for col in columns_to_remove if col in df.columns]
df = df.drop(columns=columns_to_remove)


plt.figure(figsize=(10, 6))
plt.scatter(df.columns[4:], df.loc[df['Country Name'] == 'Afghanistan'].values[0][4:], label='Afghanistan', alpha=0.5)
plt.scatter(df.columns[4:], df.loc[df['Country Name'] == 'Argentina'].values[0][4:], label='Argentina', alpha=0.5)
plt.scatter(df.columns[4:], df.loc[df['Country Name'] == 'United Arab Emirates'].values[0][4:], label='United Arab Emirates', alpha=0.5)
plt.xlabel('Ano')
plt.ylabel('Taxa de mortalidade infantil (por 1000 nascidos vivos)')
plt.title('Taxa de mortalidade infantil ao longo do tempo')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

### Conclusão

Esses gráficos nos ajudam a entender a distribuição e a relação entre os países com base em suas regiões e grupos de renda. Cada tipo de gráfico oferece uma perspectiva única que pode ser útil para diferentes tipos de análise e interpretação dos dados.
